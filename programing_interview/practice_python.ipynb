```python
# Create a link list

class Node():
    def __init__(self,data) -> None:
        self.data = data
        self.next = None


class Linklist():
    def __init__(self) -> None:
        self.head = None


    def insert(self,data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head = new_node
    def traverse(self):
        current_node = self.head
        while current_node is not None:
            print(current_node.data)
            current_node = current_node.next

    def iterative_reverse(self):
        previous_node = None
        next_node = None
        current_node = self.head

        while current_node is not None:
            # Save the current_next_node link into 
            next_node = current_node.next
            # Break the current.next node link
            current_node.next = previous_node
            # Set the previous node to the current node
            previous_node = current_node
            # Increment the current node 
            current_node = next_node
        
        # When the loop ends it will end at None so set the head node with previous node
        self.head = previous_node


        
ll_obj = Linklist()
ll_obj.insert(5)
ll_obj.insert(55)
ll_obj.insert(555)
ll_obj.traverse()
print(ll_obj.head.__dict__)
ll_obj.iterative_reverse()
print(ll_obj.head.__dict__)

```
